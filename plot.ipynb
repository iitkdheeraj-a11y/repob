```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'Month': [5, 6, 7, 8, 9, 10, 11],
        'Temp': [20, 35, 50, 30, 125,55,75]}
df = pd.DataFrame(data)

# Simple line plot
df.plot(kind='line', x='Month', y='Temp', title='Temp vs month')
plt.show()
```
